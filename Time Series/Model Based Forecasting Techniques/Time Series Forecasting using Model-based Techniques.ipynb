```python
train_data=footfalls_data.head(147)
print(train_data.shape)
test_data=footfalls_data.tail(12)
print(test_data.shape)
```

## 6. Model Training || 7. Model Testing || 8. Model Evaluation

```python
import statsmodels.formula.api as smf 
```

```python
#Linear Model
linear_model = smf.ols('Footfalls~t',data=train_data).fit()

pred_linear =  linear_model.predict(test_data['t'])

rmse_linear = np.sqrt(np.mean((np.array(test_data['Footfalls'])-np.array(pred_linear))**2))
rmse_linear
```

```python
#Exponential

exp_model = smf.ols('log_footfalls~t',data=train_data).fit()

pred_exp = exp_model.predict(test_data['t'])

rmse_exp = np.sqrt(np.mean((np.array(test_data['Footfalls'])-np.array(np.exp(pred_exp)))**2))
rmse_exp
```

```python
#Quadratic 

quad_model = smf.ols('Footfalls~t+t_square',data=train_data).fit()
pred_quad = quad_model.predict(test_data[['t','t_square']])
rmse_quad = np.sqrt(np.mean((np.array(test_data['Footfalls'])-np.array(pred_quad))**2))
rmse_quad
```

```python
#Additive seasonality 

additive_seasonality = smf.ols('Footfalls~Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov',data=train_data).fit()

pred_additive = additive_seasonality.predict(test_data[['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov']])

rmse_additive = np.sqrt(np.mean((np.array(test_data['Footfalls'])-np.array(pred_additive))**2))
rmse_additive
```

```python
#Additive Seasonality Quadratic 

additive_seasonality_quad = smf.ols('Footfalls~t+t_square+Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov',data=train_data).fit()
pred_add_quad = additive_seasonality_quad.predict(test_data[['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','t','t_square']])
rmse_add_quad = np.sqrt(np.mean((np.array(test_data['Footfalls'])-np.array(pred_add_quad))**2))
rmse_add_quad
```

```python
##Multiplicative Seasonality

multiplicative_seasonality = smf.ols('log_footfalls~Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov',data = train_data).fit()

pred_multi = multiplicative_seasonality.predict(test_data[['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov']])

rmse_multi = np.sqrt(np.mean((np.array(test_data['Footfalls'])-np.array(np.exp(pred_multi)))**2))
rmse_multi
```

```python
#Compare the results 

data = {"MODEL":["rmse_linear","rmse_exp","rmse_quad","rmse_additive_seasonality","rmse_additive_seasonality_quadratic","rmse_multiplicative"],
        "RMSE_Values":[rmse_linear,rmse_exp,rmse_quad,rmse_additive,rmse_add_quad,rmse_multi]}
table_rmse=pd.DataFrame(data)
table_rmse.sort_values(['RMSE_Values'])
```
