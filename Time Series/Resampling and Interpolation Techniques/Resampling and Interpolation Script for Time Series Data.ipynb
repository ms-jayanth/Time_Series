```python
shampoo_sales=pd.read_csv('sales.csv', index_col=0, parse_dates=True)
shampoo_sales
```

```python
plt.figure(figsize=(16,6))
plt.plot(shampoo_sales['Sales'])
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Shampoo Sales between 1990-1992')
plt.show()
```

## Resampling Technique

### 1. Up-Sampling

#### Day-wise data

```python
# Resample time-series data.
daywise_shampoo_sales=shampoo_sales.resample('D').mean()
daywise_shampoo_sales.head(45)
```

```python
#Fill NaN values using an interpolation method.
daywise_interpolated_shampoo_sales=daywise_shampoo_sales.interpolate(methos='linear')
daywise_interpolated_shampoo_sales.head(40)
```

```python
plt.figure(figsize=(16,6))
plt.plot(daywise_interpolated_shampoo_sales['Sales'])
plt.xlabel('Daywise Information')
plt.ylabel('Sales')
plt.title('Shampoo Sales between 1990-1992')
plt.show()
```

### Down-Sampling

#### Quaterly-wise data

```python
quaterlywise_shampoo_sales=shampoo_sales.resample('Q').mean()
quaterlywise_shampoo_sales.head(45)
```

```python
plt.figure(figsize=(16,6))
plt.plot(quaterlywise_shampoo_sales['Sales'])
plt.xlabel('Quaters')
plt.ylabel('Sales')
plt.title('Shampoo Sales between 1990-1992')
plt.show()
```
